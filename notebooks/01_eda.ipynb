```python
import numpy as np


colonnes_50_vide = df.columns[df.isna().mean() >= 0.5]


df[colonnes_50_vide] = df[colonnes_50_vide].astype('object').where(df[colonnes_50_vide].notna(), None)


print("Colonnes très vides :", colonnes_50_vide.tolist())

print("Nombre :", len(colonnes_50_vide))
```

```python
from collections import defaultdict
import pandas as pd


categorical_cols = [col for col in df.columns if df[col].nunique() <= 100]


value_sets = {col: set(df[col].dropna().astype(str).unique()) for col in categorical_cols}


groups = defaultdict(list)
for col, vals in value_sets.items():
    groups[frozenset(vals)].append(col)

for vals, cols in groups.items():
    print("Valeurs :", sorted([str(v) for v in vals]))
    print("Colonnes :", cols)
    print("-" * 40)

rows = []
for vals, cols in groups.items():
    for col in cols:
        rows.append({
            "colonne": col,
            "valeurs": ",".join(sorted([str(v) for v in vals]))
        })

df_groups = pd.DataFrame(rows)


df_groups = df_groups.sort_values(by="valeurs").reset_index(drop=True)

print(df_groups)
```
